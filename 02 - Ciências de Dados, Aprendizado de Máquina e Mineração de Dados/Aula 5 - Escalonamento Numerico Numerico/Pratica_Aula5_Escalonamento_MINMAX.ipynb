## **Exemplo - NORMALIZAÇÃO DE DADOS**

**OBJETIVO**:
## Técnicas de Normalização de Dados e Padronização de Dados: MinMaxScaler, StandardScaler
##

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Carregamento do dataset california_housing_train (9 atributos, 17.000 objetos)
file_path = 'california_housing_train.csv'

data = pd.read_csv(file_path).sample(100)

# Extraindo algumas colunas
rooms = data['total_rooms']
age = data['housing_median_age']
```

```python
data.info()
```

```python
# Aplicando a transformação 'Min-Max Scaling'

min_max_scaler = MinMaxScaler()
data_min_max_scaled = min_max_scaler.fit_transform(data[['total_rooms', 'housing_median_age']])
```

```python
# Exibindo os dados sem a normalização
plt.figure(figsize=(12, 10))

plt.subplot(2, 1, 1)
plt.scatter(rooms, age, color='blue', alpha=0.5)
plt.xlabel('Rooms')
plt.ylabel('Age')
plt.title('Sem Normalização')

plt.subplot(2, 1, 2)
plt.scatter(data_min_max_scaled[:, 0], data_min_max_scaled[:, 1], color='green', alpha=0.5)
plt.xlabel('Rooms (Min-Max Scaled)')
plt.ylabel('Age (Min-Max Scaled)')
plt.title('Com Normalização Min-Max Scaling')
plt.show()
```

```python
# Aplicando a normalização Z-score
z_scaler = StandardScaler()
data_z_scaled = z_scaler.fit_transform(data[['total_rooms', 'housing_median_age']])
```

```python
plt.figure(figsize=(12, 10))

plt.subplot(2, 1, 1)
plt.scatter(rooms, age, color='blue', alpha=0.5)
plt.xlabel('Rooms')
plt.ylabel('Age')
plt.title('Sem Normalização')

plt.subplot(2, 1, 2)
plt.scatter(data_z_scaled[:, 0], data_z_scaled[:, 1], color='red', alpha=0.5)
plt.xlabel('Rooms (Z-Norm)')
plt.ylabel('Age (Z-Norm)')
plt.title('Com a Normalização Z-score')
```

```python
#Calculando a média e a variância dos dados transformados
data_z_scaled_mean = data_z_scaled.mean(axis=0)
data_z_scaled_std = data_z_scaled.std(axis=0)
print("Média dos dados transformados:", data_z_scaled_mean)
print("Variância dos dados transformados:", data_z_scaled_std)
```
